Sistema di Calcolo Sconto Cliente Realizza un programma Python che simuli un sistema di calcolo sconto per un negozio. Il programma deve:

Richiedere all’utente: nome cliente 
importo totale della spesa
età del cliente

Applica uno sconto in base alle seguenti condizioni: 
Se l’importo è: 
≥ 100€ → sconto 10% 
≥ 50€ e < 100€ → sconto 5%
< 50€ → nessuno sconto 

Sconto aggiuntivo:
 Se età < 18 → +5% 
 Se età > 65 → +7%

Gli sconti sono cumulativi (somma delle percentuali)

Calcolare: sconto totale (%) valore sconto (€) prezzo finale

Stampare un riepilogo con:

 - Nome cliente
 - Importo iniziale
 -Percentuale sconto totale
 -Importo sconto
 -Prezzo finale


```python
nome = input("Nome cliente: ").strip()
importo_totale = float(int(input("Inserisci importo totale della spesa: ").strip()))
eta = int(input("Età cliente: ").strip())
sconto = 0

if importo_totale >= 100:
   sconto += 10
elif importo_totale >= 50 and importo_totale < 100:
    sconto += 5
  
if eta < 18:
  sconto += 5
elif eta > 65:
  sconto += 7

percentuale_sconto =f"{sconto}%"
valore_sconto= importo_totale * (sconto /100)
prezzo_finale = importo_totale - sconto

print(f"Nome cliente: {nome}")
print(f"Importo iniziale: {importo_totale} €")
print(f"Percentuale dello sconto: {percentuale_sconto}")
print(f"Importo sconto: {valore_sconto} €")
print(f"Prezzo finale: {prezzo_finale} €")
```

Sistema di Accesso con Tentativi Limitati Realizza un programma Python che simuli un sistema di login da terminale.

Il programma deve: Richiedere all’utente: username e password Confrontare i dati inseriti con credenziali predefinite: username_correto = "admin" password_corretta = "1234"

L’utente ha massimo 3 tentativi Ad ogni errore:

mostrare un messaggio di errore
indicare i tentativi rimanenti Se le credenziali sono corrette:
stampare messaggio di accesso consentito
terminare il ciclo Se i tentativi finiscono:
bloccare l’accesso
stampare messaggio finale
Controlli obbligatori Username vuoto → messaggio di errore Password vuota → messaggio di errore

Output atteso (esempio) Tentativo 1/3 Username: admin Password: 1111 Credenziali errate. Tentativi rimasti: 2

Tentativo 2/3 Username: admin Password: 1234 Accesso consentito!

```python
username_corretto = "admin"
password_corretta = "1234"

tentativi_massimi = 3
tentativo = 1

while tentativo <= tentativi_massimi:
    print(f"\nTentativo {tentativo}/{tentativi_massimi}")
    
    username = input("Username: ")
    password = input("Password: ")
    
    # Controllo campi vuoti
    if username == "":
        print("Errore: username non può essere vuoto.")
        continue
    
    if password == "":
        print("Errore: password non può essere vuota.")
        continue
    
    # Controllo credenziali
    if username == username_corretto and password == password_corretta:
        print("Accesso consentito!")
        break
    else:
        tentativi_rimasti = tentativi_massimi - tentativo
        if tentativi_rimasti > 0:
            print(f"Credenziali errate. Tentativi rimasti: {tentativi_rimasti}")
        else:
            print("Accesso bloccato. Hai esaurito i tentativi.")
    
    tentativo += 1
```
