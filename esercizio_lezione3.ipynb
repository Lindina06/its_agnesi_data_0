Sistema di Gestione Voti Studenti Realizza un programma Python che permetta di:

Inserire una serie di voti (numeri interi) Analizzare i voti inseriti Mostrare un riepilogo finale

Inserimento dati

Chiedere all’utente di inserire voti uno alla volta da terminale
Salvare i voti in una lista
L’inserimento termina quando l’utente digita: fine
Durante l’inserimento: Il voto deve essere:

numerico
compreso tra 0 e 30 In caso di errore:
mostrare un messaggio
NON aggiungere il valore alla lista
Elaborazione dati Utilizzando un ciclo for, calcolare:

Somma totale dei voti
Media dei voti
Numero di voti sufficienti (≥ 18)
Numero di voti insufficienti (< 18)
Voto massimo
Voto minimo
Usare if-elif-else per classificare la media:

Media ≥ 27 → "Eccellente"
Media ≥ 21 → "Buono"
Media ≥ 18 → "Sufficiente"
Media < 18 → "Insufficiente"
Output richiesto

Stampare:

Lista voti inseriti
Media
Numero sufficienti / insufficienti
Voto massimo e minimo
Classificazione finale

```python
voti = []

while True:
    valore = input('Inserisci un voto e digita: fine per terminare: ')
# Inserire una serie di voti (numeri interi)
   

# Controllo se il ciclo deve terminare
    if valore.lower() == 'fine':
        print('Fine inserimento voti.')
        break
    
    if valore.isdigit():
        voto = int(valore)
    
        if voto >= 0 and voto <= 30:
            voti.append(voto)
        else:
            print('Devi inserire un numero intero compreso tra 0 e 30.')
    else:
            print('Devi inserire un valore numerico.')
print('Inserimento voti terminato.')


somma = 0 # Somma totale dei voti
sufficienti = 0 # Contatore dei voti sufficienti (>= 18)
insufficienti = 0 # Contatore dei voti insufficienti (< 18)
massimo = voti[0] # Voto massimo all'inizio è il primo voto, poi sarà aggiornato durante l'iterazione
minimo = voti[0] # Voto minimo all'inizio è il primo voto, poi sarà aggiornato durante l'iterazione

 # Analizzare i voti inseriti            
if len(voti) == 0:
    print('Nessun voto inserito.') 
else:

    for v in voti:

        somma += v # Somma dei voti
        
        if v >= 18: sufficienti += 1 # Conto voti sufficienti
        else: insufficienti += 1 # Conto voti insufficienti
        
        if v > massimo: massimo = v # Aggiorno il voto massimo per trovare il numero massimo di voti inseriti
        if v < minimo: minimo = v # Aggiorno il voto minimo per trovare il numero minimo di voti inseriti
    media = somma / len(voti) # Calcolo la media dei voti   
    
    
if media >= 27:
    classificazione = 'Eccelente'
elif media >= 21:
    classificazione = 'Buono'
elif media >= 18:
    classificazione = 'Sufficiente'
else:
    classificazione = 'Insufficiente'
       

# Mostrare un riepilogo finale
print(voti)
print("Media voti:", media)
print("Voti sufficienti:", sufficienti)
print("Voti insufficienti:", insufficienti)
print("Voto massimo:", massimo)
print("Voto minimo:", minimo)








```
